```python
def dist3d(a, b):
    dx = a[0] - b[0]
    dy = a[1] - b[1]
    dz = a[2] - b[2]
    
    return round(sqrt(dx*dx + dy*dy + dz*dz), 2)
```

```python
def dist2d(a, b):
    dx = a[0] - b[0]
    dy = a[1] - b[1]
    
    return round(sqrt(dx*dx + dy*dy), 2)
```
