```python
def dista(a, b):
    # dist a to b
    
    dx = a[0] - b[0]
    dy = a[1] - b[1]
    dist = round(sqrt(dx*dx + dy*dy), 2)    

    return dist
```

```python
def fahrestone(a, tk_dict):
    # return fahrest point to a from c_list
    # list of tk (set, x, y)
    dist_tk = 0
    fahrest_tk = (0, 0)
    
    if len(tk_dict) == 6:
        # find one middle point for pair
        
        sets = sorted(tk_dict.keys())  # names of sets
        #print(sets)
        new_dict = {}
        for k in range(3):
            #print(k, sets[k], tk_dict[sets[k]][0], tk_dict[sets[k]][1])
            new_dict[sets[k]] = (((tk_dict[sets[k]][0] + tk_dict[sets[k]+10][0])/2), ((tk_dict[sets[k]][1] + tk_dict[sets[k]+10][1])/2))
        
        tk_dict = new_dict
    
    #check dist
    for n in tk_dict.values():
        if dista(a, (n[0], n[1])) > dist_tk:
            dist_tk = dista(a, (n[0], n[1]))
            fahrest_tk = (n[0], n[1])
    
    return fahrest_tk
```

```python
# создаем словарь и заносим туда тк для каждой опоры
# zones = {}    # (ctow, left tk, right tk)

left_sets = [1, 2, 3, 11, 12, 13, '1', '2', '3', '11', '12', '13']
right_sets = [4, 5, 6, 14, 15, 16, '4', '5', '6', '14', '15', '16']

for n in range(len(cgt_tab)):
    idx = cgt_tab.iloc[n].name
    struct_coo = (cgt_tab.loc[idx, 'x'], cgt_tab.loc[idx, 'y'])
    
    # create dicts for left and right tk and fill them
    ltk_dict = {}    
    rtk_dict = {}
    
    for tk in range(len(tk_tab)):
        if tk_tab.loc[tk, 'id'] == idx:
            if tk_tab.loc[tk, 'set'] in left_sets:
                ltk_dict[tk_tab.loc[tk, 'set']] = (tk_tab.loc[tk, 'x'], tk_tab.loc[tk, 'y'])

            elif tk_tab.loc[tk, 'set'] in right_sets:
                rtk_dict[tk_tab.loc[tk, 'set']] = (tk_tab.loc[tk, 'x'], tk_tab.loc[tk, 'y'])
            
    # choose one fahrest coord from list
    cgt_tab.loc[idx, 'ltk_x'] = fahrestone(struct_coo, ltk_dict)[0]
    cgt_tab.loc[idx, 'ltk_y'] = fahrestone(struct_coo, ltk_dict)[1]
    cgt_tab.loc[idx, 'rtk_x'] = fahrestone(struct_coo, rtk_dict)[0]
    cgt_tab.loc[idx, 'rtk_y'] = fahrestone(struct_coo, rtk_dict)[1]

```

```python
# then we need to find lines and build parallel linesof them

# add new cols
cgt_tab.assign(area_c=np.nan, area=np.nan)
list_of_area = []

for n in range(len(cgt_tab)-1):
    idx = cgt_tab.iloc[n].name
    left_line = LineString([(cgt_tab.iloc[n].ltk_x, cgt_tab.iloc[n].ltk_y), (cgt_tab.iloc[n+1].ltk_x, cgt_tab.iloc[n+1].ltk_y)])
    right_line = LineString([(cgt_tab.iloc[n].rtk_x, cgt_tab.iloc[n].rtk_y), (cgt_tab.iloc[n+1].rtk_x, cgt_tab.iloc[n+1].rtk_y)])

    left_paral = left_line.parallel_offset(zona_size, 'left')      # left parallel line
    right_paral = right_line.parallel_offset(zona_size, 'right')      # right parallel line
    
    zona = Polygon([(cgt_tab.iloc[n].x, cgt_tab.iloc[n].y), 
        list(left_paral.coords)[0], 
        list(left_paral.coords)[1],
        (cgt_tab.iloc[n+1].x, cgt_tab.iloc[n+1].y),
        list(right_paral.coords)[0],
        list(right_paral.coords)[1]])
    
    print(zona.area)
    
    list_of_area.append(zona)
    cgt_tab.loc[idx, 'area_c'] = str(list(zona.exterior.coords))
    cgt_tab.loc[idx, 'area'] = zona.area


    
```
