парсинг файлов опор для вычленения оттуда углов наклона осей

```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
# пути 
workdir = Path(r'D:\work\2022_m\110_man-nps1\test\angles')
strdir = Path(r'D:\work\2022_m\110_man-nps1\test\angles\Stick')
```

```python
# read names
cgt_tab = pd.read_excel(workdir / 'all_in_one.xlsx', index_col=0, names=['id', 'path'], sheet_name='str_names')
```

```python
cgt_tab
```

```python
# add new cols
cgt_tab.assign(dx=np.nan, dy=np.nan)

for stk in range(len(cgt_tab)):
    idx = cgt_tab.iloc[stk].name
    with open(strdir / cgt_tab.iloc[stk].path) as str_file:
        lines = str_file.readlines()
        cgt_tab.loc[idx, 'dx'] = -float(lines[11].split()[0])
        cgt_tab.loc[idx, 'dy'] = float(lines[11].split()[1])
        #print(lines[11].split()[0], lines[11].split()[1])

```

```python
cgt_tab
```

```python
# save tab
cgt_tab.to_csv(workdir / 'lean_angles.txt', sep='\t', encoding = 'cp1251')
```
